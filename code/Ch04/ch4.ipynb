```python
# read the data/broadcast_logs folder
import os
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = SparkSession.builder.appName("ch4").getOrCreate()

DIRECTORY = "/home/nyck33/Documents/DataEngineering/DataAnalysisWithPythonAndPySpark/code/data/broadcast_logs/"
logs = spark.read.csv(
    path=os.path.join(DIRECTORY, "BroadcastLogs_2018_Q3_M8_sample.CSV"), 
    sep="|", 
    header=True, 
    inferSchema=True, 
    timestampFormat="yyyy-MM-dd",
    )
```

```python
# sample_frame.columns # => ['item', 'price', 'quantity', 'UPC']
# Create an empty dataframe with the same columns as sample_frame comment given as guidance
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType
sample_schema = StructType([
    StructField("item", StringType(), True),
    StructField("price", DoubleType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("UPC", StringType(), True)

])

sample_frame = spark.createDataFrame([], schema=sample_schema)

print(sample_frame.columns) # => ['item', 'price', 'quantity', 'UPC']
# Drop the columns 'item', 'UPC', and 'prices' from the dataframe
print(sample_frame.drop('item', 'UPC', 'prices').columns) # => ['quantity']


```

```python
logs.select(F.col("Duration")).show(5)
print(logs.select(F.col("Duration")).dtypes)
```

```python
logs.select(
    F.col("Duration"),
    F.col("Duration").substr(1,2).cast("int").alias("dur_hours"),
    F.col("Duration").substr(4,2).cast("int").alias("dur_minutes"),
    F.col("Duration").substr(7,2).cast("int").alias("dur_seconds"),
).distinct().show(
    5
)
```

```python
# peform arithmetic to get duration in seconds
logs.select(
    F.col("Duration"),
    F.col("Duration").substr(1,2).cast("int") * 60 * 60 + \
    F.col("Duration").substr(4,2).cast("int") * 60 + \
    F.col("Duration").substr(7,2).cast("int").alias("dur_seconds"),
).distinct().limit(5).toPandas() # better formatting

```

```python
logs = logs.withColumn(
    "Duration_seconds",
    (
        F.col("Duration").substr(1,2).cast("int") * 60 * 60 + \
        F.col("Duration").substr(4,2).cast("int") * 60 + \
        F.col("Duration").substr(7,2).cast("int")
    ),
)

logs.printSchema()

```

```python
logs = logs.withColumnRenamed("Duration_seconds", "duration_seconds")

logs.printSchema()
```
